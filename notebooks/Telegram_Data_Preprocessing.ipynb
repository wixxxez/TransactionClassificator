```python
import pandas as pd
import json
from datetime import datetime
from tqdm.notebook import tqdm
```

```python
telegram_df=pd.read_json("../data/TelegramHistory.json")
```

```python
message_history = telegram_df['messages'].to_list()
```

```python

message_history_parts = []
for message in tqdm(message_history[17:]): 

    timestamp = message['date_unixtime']

     
    try: 
        message_body = message['text'].replace("\n",':').split(":") 

        amount = message_body[5]
        amount = int(float(amount.strip()))
        description = message_body[11]
        


        date_time = datetime.fromtimestamp(int(timestamp))
    
        full_date = date_time.strftime("%d.%m.%Y")
        month_number = date_time.month
        week_number = date_time.isocalendar()[1]


        message_history_parts.append(pd.DataFrame({
            'time' : timestamp,
            'description': description,
            'amount' : amount,
            'full_date' : full_date,
            'month_number' : month_number,
            'week_number': week_number,
        }, index=['A']))
    except AttributeError as e :

        print(message['text'])

        
    
    
```

```python
telegram_df_messages = pd.concat(message_history_parts).reset_index(drop=True)
telegram_df_messages
```

```python
telegram_df_messages.to_csv("../data/MonoTransactionFromTelegram_ARTEM.csv")
```
